```python
import glob,os

path = '/Users/corinnsmall/Documents/BPDCN/bpdcn_papers/single_case_papers/'
files = glob.glob(path + '*.pdf')

for file in files:
    name = file.split('/')[-1]
    print(name)
    if 'copy' in name:
        parsed = name.split(' copy')
        #print(parsed)
        new_name = ''.join(parsed)
        print(new_name)
        
        
        os.rename(file, path + new_name)
        print('renamed')
    
```
